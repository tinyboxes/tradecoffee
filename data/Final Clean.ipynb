```python
import pandas as pd
import numpy as np
```

```python
dfx = pd.read_csv('./extra_clean.csv', index_col=0)
```

```python
dfx[['extra', 'process', 'sub region', 'varietal','harvest time','elevation', 'producer']] = dfx[['extra', 'process', 'sub region', 'varietal','harvest time','elevation', 'producer']].astype(str)
```

```python
dfx[:1]
```

```python
dfx['extra'] = dfx['extra'].map(lambda x: x.lstrip('[').rstrip(']'))
dfx['extra'] = dfx['extra'].map(lambda x: x.replace(',', '').replace('\'', ''))
```

```python
dfx['process'] = dfx['process'].map(lambda x: x.replace(',', '').replace('\'', ''))
dfx['process'] = dfx['process'].map(lambda x: x.lstrip('[').rstrip(']'))
```

```python
dfx['sub region'] = dfx['sub region'].map(lambda x: x.replace(',', '').replace('\'', ''))
dfx['sub region'] = dfx['sub region'].map(lambda x: x.lstrip('[').rstrip(']'))
```

```python
dfx['varietal'] = dfx['varietal'].map(lambda x: x.replace(',', '').replace('\'', ''))
dfx['varietal'] = dfx['varietal'].map(lambda x: x.lstrip('[').rstrip(']'))
```

```python
dfx['elevation'] = dfx['elevation'].map(lambda x: x.replace(',', '').replace('\'', ''))
dfx['elevation'] = dfx['elevation'].map(lambda x: x.lstrip('[').rstrip(']'))
```

```python
dfx[dfx['sub region'] == 'nan'] = np.nan
```

```python
x = ['process','sub region', 'varietal','harvest time','elevation','producer']

dfx[dfx[x]== 'nan'] = np.nan
```
