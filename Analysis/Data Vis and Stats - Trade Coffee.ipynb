Created a new data frame to work with product price data.

```python
price_cup = df[['name', 'price','weight in g', 'blend_single', 'classification','roaster']]
```

Using my own pour over recipe for a cup of coffee wherein I use 20 g of beans for 300 mL of water. 1:15 bean to water ratio.

By dividing the total weight of beans by 20 g, I get the number of cups of coffee per bag.

```python
price_cup['cups'] = df['weight in g']/20

# 20 g of coffee to make 300 ml of coffee
```

By dividing the price of each bag by the number of cups, I get the price per cup of coffee, rounded to the first 2 decimal places.

```python
price_cup['ppc'] = price_cup['price']/price_cup['cups']
```

```python
price_cup['rounded'] = price_cup['ppc'].values.round(2)
```
